# HTML
HTML - Hyper Text Markup Language (https://www.w3schools.com/html/html_intro.asp)

HTML элементы описываются тегами

```
<b>Жирный текст</b>
```
У тегов есть аттрибуты

```
<a href="http://ya.ru">Яндекс</a>
```

Теги могут вкладываться друг в друга
```
<div id="main">
    <div class="sub">
        <a href="http://yandex.ru">Яндекс</a>
    </div>
    <div class="sub">
        <a href="http://google.com">Google</a>
    </div>
    </div>
<div id="footer">
</div>
```

К html элементам можно обращаться по имени
```
a
```
По идентификатору
```
#main
```
По классу
```
.sub
```
Вложенность
```
#main .sub
```



````python
!pip install beautifulsoup4

html_doc = """
<html><head><title>Тестовый html-документ</title></head>
<body>
<div id="main">
<div class="sub">
<a href="http://yandex.ru">Яндекс</a>
</div>
<div class="sub">
<a href="http://google.com">Google</a>
</div>
</div>
<div id="footer">
</div>
</body>
</html>
"""

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')

soup.select('a')
````

````python
all_topics = []

for i in range(2, 11):
    print('Загружаю страницу', i)
    
    link = 'https://www.hse.ru/sci/diss/index' + str(i) + '.html'
    page = urllib.request.urlopen('https://www.hse.ru/sci/diss/').read()
    soup = BeautifulSoup(page, 'html.parser')
    
    topics = soup.select('h2.dissertation_title a')
    
    for topic in topics:
        all_topics.append(topic.text)
        
print('\n'.join(all_topics))
````
