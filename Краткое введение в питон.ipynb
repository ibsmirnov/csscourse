__Цикл for__

Используется чаще, чем while.
Отличие от while:
-  while выполняет цикл (итерацию), пока выполяется условие
-  for выполняет цикл, пока не кончатся элементы, с которыми нужно что-то сделать. 

```python
# Цикл for:

mylist1 = [1, 2, 3, 4]

for i in mylist1:  
    i += 1
    print(i) # берем элементы списка по очереди, прибавляем к ним 1, но сам список не меняется

    # исходный список не меняется, потому что python в каждой итерации присваивает элементу списка имя i.
    # и дальше мы меняем значение i -- а не сам элемент списка
print(mylist1)
```

```python
# чаще удобнее обращаться по индексу, а не перебирать сами элементы. С помощью обращений по индексу мы можем изменить список
# len(mylist1) -- возвращает длину списка. 
# функция range(x, y) создает последовательность чисел от x до y, правая граница не включается
# здесь мы изменили список 

for i in range(0, len(mylist1)): 
    mylist1[i] += 1
print(mylist1)
```

For удобнее во многих случаях, чем while, т.к. не нужен счетчик. И for работает быстрее. Простое правило: если есть четкий набор элементов, используйте for.

__Словарь (dictionary)__
Словарь -- неупорядоченные элементы, к которым можно обращаться по ключу. Ключи уникальны. В словари удобно сохранять разную информацию по одному и тому же объекту.

```python
mydict1 = {'Саша': 41, 'Света': 32, 'Маша': 49} # 'Саша', Света, Маша -- ключи, а 41, 32, 49 значения
print(mydict1['Саша'])

mydict2 = {'Саша': [41, 'male'], 'Света': [32, 'female'], 'Маша': [49, 'female']} 
mydict2['Саша']
```

```python
# Давайте сделаем словарь, в который для каждого человека (user_id) сохраним среднюю успеваемость в школе и пол
user_id = ['1519_1', '1519_2', '1519_3']
gpa = ['267', '179', '299']
gender = ['f', 'f', 'm']
data_1 = {}
for i in range(0, len(user_id)):
    data_1[user_id[i]] = [gender[i], gpa[i]]
print(data_1)

```

```python
# Давайте посчитаем частоту разных слов, сделав словарь:
passage = 'Тут какая-то тестовая строка строка строка'
words = {}
for word in passage.split():
    if word not in words: # если слова(ключа) пока нет в словаре -- давайте его добавим и посчитаем
        words[word] = 1      
    else:        
        words[word] += 1 # а если слово уже было -- давайте обновим значение, т.к. слово встретилось еще раз
print(words)
```

А теперь немного о функциях. int(), str(), print() -- все это примеры функций. Функции можно задавать и самостоятельно.
